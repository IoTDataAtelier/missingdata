<h2>Loading dataset</h2>

```python
from pypots.benchpots.datasets import preprocess_physionet2012
physionet2012_dataset = preprocess_physionet2012(subset="all", rate=0.1)
```

<h3>Training data</h3>

<h4>Loading training dataset</h4>

```python
train_X = physionet2012_dataset['train_X']
```

```python
female_gender_training = train_X[train_X["Gender"] == 0.0]
female_gender_training = female_gender_training["RecordID"]
female_gender_distribution_training = train_X[train_X["RecordID"].isin(female_gender_training)]
female_gender_distribution_training = female_gender_distribution_training.count()
female_gender_distribution_training
```

```python
male_gender_training = train_X[train_X['Gender'] == 1.0]
male_gender_training = male_gender_training["RecordID"]
male_gender_distribution_training = train_X[train_X["RecordID"].isin(male_gender_training)]
male_gender_distribution_training = male_gender_distribution_training.count()
male_gender_distribution_training 
```

```python
undefined_gender_training = train_X[train_X['Gender'] == -1.0]
undefined_gender_training = undefined_gender_training['RecordID']
undefined_gender_distribution_training = train_X[train_X["RecordID"].isin(undefined_gender_training)]
undefined_gender_distribution_training = undefined_gender_distribution_training.count()
undefined_gender_distribution_training
```

```python
ICUType_1_training = train_X[train_X['ICUType'] == 1.0]
ICUType_1_training = ICUType_1_training[ICUType_1_training["Time"] == 0.0]
ICUType_1_training= ICUType_1_training["RecordID"]
ICUType_1_distribution_training = train_X[train_X["RecordID"].isin(ICUType_1_training)]
ICUType_1_distribution_training = ICUType_1_distribution_training.count() 
```

```python
ICUType_2_training = train_X[train_X['ICUType'] == 2.0]
ICUType_2_training = ICUType_2_training[ICUType_2_training["Time"] == 0.0]
ICUType_2_training= ICUType_2_training["RecordID"]
ICUType_2_distribution_training = train_X[train_X["RecordID"].isin(ICUType_2_training)]
ICUType_2_distribution_training = ICUType_2_distribution_training.count() 
```

```python
ICUType_3_training = train_X[train_X['ICUType'] == 3.0]
ICUType_3_training = ICUType_3_training[ICUType_3_training["Time"] == 0.0]
ICUType_3_training= ICUType_3_training["RecordID"]
ICUType_3_distribution_training = train_X[train_X["RecordID"].isin(ICUType_3_training)]
ICUType_3_distribution_training = ICUType_3_distribution_training.count() 
```

```python
ICUType_4_training = train_X[train_X['ICUType'] == 4.0]
ICUType_4_training = ICUType_4_training[ICUType_4_training["Time"] == 0.0]
ICUType_4_training= ICUType_4_training["RecordID"]
ICUType_4_distribution_training = train_X[train_X["RecordID"].isin(ICUType_4_training)]
ICUType_4_distribution_training = ICUType_4_distribution_training.count() 
ICUType_4_distribution_training 
```

```python
more_than_or_equal_to_65_training = train_X[train_X["Age"] >= 65]
more_than_or_equal_to_65_training = more_than_or_equal_to_65_training[more_than_or_equal_to_65_training["Time"] == 0.0]
more_than_or_equal_to_65_training = more_than_or_equal_to_65_training["RecordID"]
more_than_or_equal_to_65_training = train_X[train_X["RecordID"].isin(more_than_or_equal_to_65_training)]
more_than_or_equal_to_65_training = more_than_or_equal_to_65_training.count()
```

```python
# Motivação para o teste: A quantidade de valores no grupo de idade que estava na tabela, era muito maior do que a quantidade de 
# valores para o grupo de gênero, o que não fazia sentido, pois, já que não temos dados faltantes dessas variáveis em ambos os grupos
# se gênero tem n ocorrências, idade deve ter n ocorrências também, e vice-versa. 

# Descoberta: Através do teste, vimos que tinha casos em que o valor do campo idade, se repete durante a série temporal do paciente, e visto isso, nós ajustamos o código para pegar apenas a primeira ocorrência de valor para cada recordID.
# Também descobrimos que, como estamos filtrando apenas os gêneros female e male e há um campo para gênero indefinido, o grupo de Age estava pegando as idades desse campo de gênero indefinido também, o que estava fazendo com que 
# a soma total de ocorrências para o grupo de Age estivesse maior do que para o grupo de Gender.

# teste2 = train_X[train_X['Age'] >= 65]
# teste2["RecordID"].value_counts()   
```

```python
less_than_or_equal_to_65_training = train_X[train_X["Age"] < 65]
less_than_or_equal_to_65_training = less_than_or_equal_to_65_training[less_than_or_equal_to_65_training["Time"] == 0.0]
less_than_or_equal_to_65_training = less_than_or_equal_to_65_training["RecordID"]
less_than_or_equal_to_65_training = train_X[train_X["RecordID"].isin(less_than_or_equal_to_65_training)]
less_than_or_equal_to_65_training = less_than_or_equal_to_65_training.count()
less_than_or_equal_to_65_training
```

```python
filtered_train_X = train_X[(train_X['Height'] != -1) & (train_X['Weight'] != -1) & (train_X['Height'].notna()) & (train_X['Weight'].notna())] 
```

```python
def classify_BMI(BMI):
    if BMI <= 18.5:
        return "Baixo peso"
    elif BMI >= 18.6 and BMI <= 24.9:
        return "Peso normal"
    elif BMI >= 25 and BMI <= 29.9:
        return "Sobrepeso"
    elif BMI >= 30 and BMI <= 34.9:
        return "Obesidade grau 1"
    elif BMI >= 35 and BMI <= 39.9:
        return "Obesidade grau 2"
    elif BMI >= 40:
        return "Obesidade grau 3"
```

```python
filtered_train_X_meters = filtered_train_X.copy()
filtered_train_X_meters['Height'] = filtered_train_X['Height'] / 100 # Converting Height from cm to meters
filtered_train_X_meters['Height']
```

```python
bmi_data_train = filtered_train_X_meters
bmi_data_train["BMI"] = round(filtered_train_X_meters["Weight"] / (filtered_train_X_meters["Height"] ** 2), 1)
bmi_data_train["Classificacao"] = bmi_data_train["BMI"].apply(classify_BMI)
bmi_data_train.head()
```

```python
bmi_data_train = bmi_data_train.groupby("RecordID").first().reset_index()
bmi_data_train
```

```python
classification_low_weight_train = bmi_data_train[bmi_data_train["Classificacao"] == 'Baixo peso']
classification_low_weight_train = classification_low_weight_train["RecordID"]
classification_l_w_t = train_X[train_X["RecordID"].isin(classification_low_weight_train)]
classification_l_w_t = classification_l_w_t.count() 
classification_l_w_t 
```

```python
classification_normal_weight_train = bmi_data_train[bmi_data_train["Classificacao"] == 'Peso normal']
classification_normal_weight_train = classification_normal_weight_train["RecordID"]
classification_n_w_t = train_X[train_X["RecordID"].isin(classification_normal_weight_train)]
classification_n_w_t = classification_n_w_t.count() 
classification_n_w_t 
```

```python
classification_overweight_train = bmi_data_train[bmi_data_train["Classificacao"] == 'Sobrepeso']
classification_overweight_train = classification_overweight_train['RecordID']
classification_o_w_t = train_X[train_X["RecordID"].isin(classification_overweight_train)]
classification_o_w_t = classification_o_w_t.count() 
classification_o_w_t 
```

```python
classification_obesity_grade1_train = bmi_data_train[bmi_data_train["Classificacao"] == 'Obesidade grau 1']
classification_obesity_grade1_train = classification_obesity_grade1_train["RecordID"]
classification_ob1_t = train_X[train_X["RecordID"].isin(classification_obesity_grade1_train)]
classification_ob1_t = classification_ob1_t.count() 
classification_ob1_t 
```

```python
classification_obesity_grade2_train = bmi_data_train[bmi_data_train["Classificacao"] == 'Obesidade grau 2']
classification_obesity_grade2_train = classification_obesity_grade2_train["RecordID"]
classification_ob2_t = train_X[train_X["RecordID"].isin(classification_obesity_grade2_train)]
classification_ob2_t = classification_ob2_t.count() 
classification_ob2_t 
```

```python
classification_obesity_grade3_train = bmi_data_train[bmi_data_train["Classificacao"] == 'Obesidade grau 3']
classification_obesity_grade3_train = classification_obesity_grade3_train["RecordID"]
classification_ob3_t = train_X[train_X["RecordID"].isin(classification_obesity_grade3_train)]
classification_ob3_t = classification_ob3_t.count() 
classification_ob3_t 
```

```python
df_columns = train_X.columns
df_columns
```

```python
df_test = pd.DataFrame(columns=df_columns)
df_test_transpose = df_test.T
df_test_transpose["Female"] = female_gender_distribution_training
df_test_transpose["Male"] = male_gender_distribution_training
df_test_transpose["Undefined Gender"] = undefined_gender_distribution_training
df_test_transpose["ICUType 1"] = ICUType_1_distribution_training 
df_test_transpose["ICUType 2"] = ICUType_2_distribution_training 
df_test_transpose["ICUType 3"] = ICUType_3_distribution_training 
df_test_transpose["ICUType 4"] = ICUType_4_distribution_training 
df_test_transpose["Age 65+"] = more_than_or_equal_to_65_training
df_test_transpose["Age 65-"] = less_than_or_equal_to_65_training
df_test_transpose['Low Weight'] = classification_l_w_t 
df_test_transpose['Normal Weight'] = classification_n_w_t 
df_test_transpose['Overweight'] = classification_o_w_t 
df_test_transpose['Obesity Grade 1'] = classification_ob1_t 
df_test_transpose['Obesity Grade 2'] = classification_ob2_t 
df_test_transpose['Obesity Grade 3'] = classification_ob3_t 

#df_test_transpose.style.set_caption("Repeated Measurements per variable by demographics")

display(HTML("<h2 style='text-align: center; font-size: 24px; font-weight: bold;'>Repeated Measurements per Variable by Demographics - Train Set</h2>"))
df_test_transpose
```
