<h2>Loading dataset</h2>

```python
from pypots.benchpots.datasets import preprocess_physionet2012
physionet2012_dataset = preprocess_physionet2012(subset="all", rate=0.1)
```

<h3>Training data</h3>

<h4>Loading training dataset</h4>

```python
train_X = physionet2012_dataset['train_X']
```

```python
female_gender_distribution_training = train_X[train_X['Gender'] == 0.0].count()
female_gender_distribution_training
```

```python
male_gender_distribution_training = train_X[train_X['Gender'] == 1.0].count()
male_gender_distribution_training
```

```python
ICUType_1_training = train_X[train_X['ICUType'] == 1.0].count() 
ICUType_1_training
```

```python
ICUType_2_training = train_X[train_X['ICUType'] == 2.0].count() 
ICUType_2_training
```

```python
ICUType_3_training = train_X[train_X['ICUType'] == 3.0].count() 
ICUType_3_training
```

```python
ICUType_4_training = train_X[train_X['ICUType'] == 4.0].count() 
ICUType_4_training
```

```python
more_than_or_equal_to_65_training = train_X[train_X['Age'] >= 65].count()
more_than_or_equal_to_65_training
```

```python
less_than_or_equal_to_65_training = train_X[train_X['Age'] < 65].count()
less_than_or_equal_to_65_training
```

```python
filtered_train_X = train_X[(train_X['Height'] != -1) & (train_X['Weight'] != -1) & (train_X['Height'].notna()) & (train_X['Weight'].notna())] 
```

```python
def classify_BMI(BMI):
    if BMI < 18.5:
        return "Baixo peso"
    elif BMI >= 18.6 and BMI <= 24.9:
        return "Peso normal"
    elif BMI >= 25 and BMI <= 29.9:
        return "Sobrepeso"
    elif BMI >= 30 and BMI <= 34.9:
        return "Obesidade grau 1"
    elif BMI >= 35 and BMI <= 39.9:
        return "Obesidade grau 2"
    elif BMI >= 40:
        return "Obesidade grau 3"
```

```python
filtered_train_X_meters = filtered_train_X.copy()
filtered_train_X_meters['Height'] = filtered_train_X['Height'] / 100 # Converting Height from cm to meters
filtered_train_X_meters['Height']
```

```python
bmi_data_train = filtered_train_X_meters
bmi_data_train["BMI"] = filtered_train_X_meters["Weight"] / (filtered_train_X_meters["Height"] ** 2)
bmi_data_train["Classificacao"] = bmi_data_train["BMI"].apply(classify_BMI)
bmi_data_train.head()
```

```python
classification_low_weight_train = bmi_data_train[bmi_data_train["Classificacao"] == 'Baixo peso']
classification_l_w_t = classification_low_weight_train.count() 
classification_l_w_t 
```

```python
classification_normal_weight_train = bmi_data_train[bmi_data_train["Classificacao"] == 'Peso normal']
classification_n_w_t = classification_normal_weight_train.count() 
classification_n_w_t 
```

```python
classification_overweight_train = bmi_data_train[bmi_data_train["Classificacao"] == 'Sobrepeso']
classification_o_w_t = classification_overweight_train.count() 
classification_o_w_t 
```

```python
classification_obesity_grade1_train = bmi_data_train[bmi_data_train["Classificacao"] == 'Obesidade grau 1']
classification_ob1_t = classification_obesity_grade1_train.count() 
classification_ob1_t 
```

```python
classification_obesity_grade2_train = bmi_data_train[bmi_data_train["Classificacao"] == 'Obesidade grau 2']
classification_ob2_t = classification_obesity_grade2_train.count() 
classification_ob2_t 
```

```python
classification_obesity_grade3_train = bmi_data_train[bmi_data_train["Classificacao"] == 'Obesidade grau 3']
classification_ob3_t = classification_obesity_grade3_train.count() 
classification_ob3_t 
```

```python
df_columns = train_X.columns
df_columns
```

```python
df_test = pd.DataFrame(columns=df_columns)
df_test_transpose = df_test.T
df_test_transpose["Female"] = female_gender_distribution_training
df_test_transpose["Male"] = male_gender_distribution_training
df_test_transpose["ICUType 1"] = ICUType_1_training
df_test_transpose["ICUType 2"] = ICUType_2_training
df_test_transpose["ICUType 3"] = ICUType_3_training
df_test_transpose["ICUType 4"] = ICUType_4_training
df_test_transpose["Age 65+"] = more_than_or_equal_to_65_training
df_test_transpose["Age 65-"] = less_than_or_equal_to_65_training
df_test_transpose['Low Weight'] = classification_l_w_t 
df_test_transpose['Normal Weight'] = classification_n_w_t 
df_test_transpose['Overweight'] = classification_o_w_t 
df_test_transpose['Obesity Grade 1'] = classification_ob1_t 
df_test_transpose['Obesity Grade 2'] = classification_ob2_t 
df_test_transpose['Obesity Grade 3'] = classification_ob3_t 

#df_test_transpose.style.set_caption("Repeated Measurements per variable by demographics")

display(HTML("<h2 style='text-align: center; font-size: 24px; font-weight: bold;'>Repeated Measurements per variable by demographics</h2>"))
df_test_transpose
```
